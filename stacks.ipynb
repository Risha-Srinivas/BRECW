```python
class Stack:
    def __init__(self):
        self.items = [] # Underlying storage
    def push(self, item):
        # .append() adds to the end (The Top)
        self.items.append(item)
    def pop(self):
        # .pop() with no arguments removes the last element
        if not self.is_empty():
            return self.items.pop()
    def peek(self):
        # Negative indexing [-1] gets the top/last element
        return self.items[-1]
    def is_empty(self):
        return len(self.items) == 0
```

```python
# Demonstrate using the Stack object
my_stack = Stack()
print(f"Is the stack empty? {my_stack.is_empty()}")

my_stack.push(10)
my_stack.push('hello')
my_stack.push(True)

print(f"After pushing items, is the stack empty? {my_stack.is_empty()}")
print(f"Top element (peek): {my_stack.peek()}")
print(f"Popping an item: {my_stack.pop()}")
print(f"New top element (peek): {my_stack.peek()}")
print(f"Popping another item: {my_stack.pop()}")
print(f"Popping the last item: {my_stack.pop()}")
print(f"Is the stack empty now? {my_stack.is_empty()}")
```

```python
#1.check parenthesis.Given an array check - Take the string [( text )]. We see [. We see (. The next closing bracket must match the ( because it was the last one opened.
def check_parentheses(text_string):
    stack = []

    for character in text_string:
        # Step 1: If it's an opening bracket, save it on the stack
        if character in ['(', '{', '[']:
            stack.append(character)

        # Step 2: If it's a closing bracket, verify it matches the top
        elif character in [')', '}', ']']:

            # Critical Edge Case: If the stack is empty but we see a closing bracket,
            # it means there was never an opening bracket for it (e.g., string is "}")
            if len(stack) == 0:
                return False

            # Pop the top element to compare it
            top_item = stack.pop()

            # Check for structural mismatches
            if character == ')' and top_item != '(':
                return False
            if character == '}' and top_item != '{':
                return False
            if character == ']' and top_item != '[':
                return False

    # Step 3: Final Check. If the stack still has items left,
    # it means some brackets were opened but never closed (e.g., string is "([]")
    if len(stack) == 0:
        return True
    else:
        return False

# Execution Examples:
print(check_parentheses("()[]{}")) # True
print(check_parentheses("([)]"))   # False
print(check_parentheses("(["))     # False (Stack isn't empty at the end)
print(check_parentheses("]"))      # False (Stack was empty when closing arrived)
```

```python
#2.implement queue using stack
# Logic for using two stacks to implement a queue:
# A queue is First-In, First-Out (FIFO). A stack is Last-In, First-Out (LIFO).
# To achieve FIFO behavior with two LIFO stacks:
# 1. 'input_stack': This stack is used for 'enqueue' operations. New items are always pushed onto this stack.
#    It temporarily holds elements in reverse order of their arrival (LIFO).
# 2. 'output_stack': This stack is used for 'dequeue' and 'get_front' operations.
#    When the 'output_stack' is empty and an item needs to be dequeued/peeked, all items from the 'input_stack'
#    are popped and pushed onto the 'output_stack'. This effectively reverses the order twice, making the oldest item
#    (the one that arrived first) available at the top of the 'output_stack'.
# This way, 'input_stack' is for receiving new elements, and 'output_stack' is for releasing the oldest elements.
class CustomQueue:
    def __init__(self):
        # Initialize our two distinct internal stacks using Python lists
        self.input_stack = []  # Used for enqueuing new elements
        self.output_stack = [] # Used for dequeuing/peeking elements (stores them in correct FIFO order)

    def enqueue(self, item):
        """Adds an item to the back of the queue."""
        # We always just drop new items into the input stack.
        # This maintains the order of arrival within the input_stack.
        self.input_stack.append(item) # Appends the new item to the end of the input_stack

    def dequeue(self):
        """Removes and returns the item at the front of the queue."""
        # If both stacks are completely empty, the queue is empty. No items to dequeue.
        if len(self.input_stack) == 0 and len(self.output_stack) == 0:
            return "Queue is empty"

        # If our output stack is empty, we must invert our input stack into it
        # This process transfers all elements from input_stack to output_stack,
        # reversing their order so the oldest element is now at the top of output_stack.
        if len(self.output_stack) == 0:
            while len(self.input_stack) > 0:
                extracted_item = self.input_stack.pop() # Remove from input_stack (LIFO)
                self.output_stack.append(extracted_item) # Add to output_stack (effectively reversing)

        # Now that the output_stack is populated (or was already), we can pop the front item.
        # This item is the one that has been in the queue the longest (FIFO).
        return self.output_stack.pop()

    def get_front(self):
        """Looks at the front item without removing it."""
        # If both stacks are completely empty, the queue is empty. No items to peek.
        if len(self.input_stack) == 0 and len(self.output_stack) == 0:
            return "Queue is empty"

        # If our output stack is empty, we must invert our input stack into it
        # Similar to dequeue, this prepares the output_stack to have the oldest item at its top.
        if len(self.output_stack) == 0:
            while len(self.input_stack) > 0:
                self.output_stack.append(self.input_stack.pop())

        # Return the top item of output_stack without popping it.
        # This is the oldest element in the queue.
        return self.output_stack[-1]

    def is_empty(self):
        """Checks if the queue has elements."""
        # The queue is empty only if both its internal stacks are empty.
        return len(self.input_stack) == 0 and len(self.output_stack) == 0

# --- Testing the Implementation ---:
print("\n--- Testing CustomQueue Implementation ---")
my_queue = CustomQueue()
print(f"Is queue empty initially? {my_queue.is_empty()}") # Expected: True

my_queue.enqueue("A")
my_queue.enqueue("B")
my_queue.enqueue("C")
print(f"Queue after enqueuing A, B, C. Is empty? {my_queue.is_empty()}") # Expected: False
print(f"Front item (peek): {my_queue.get_front()}") # Expected: A

print(f"Dequeued: {my_queue.dequeue()}")  # Expected: A (First In, First Out)
print(f"Front item (peek): {my_queue.get_front()}") # Expected: B

my_queue.enqueue("D")
print(f"Queue after enqueuing D. Front item: {my_queue.get_front()}") # Expected: B

print(f"Dequeued: {my_queue.dequeue()}")  # Expected: B
print(f"Dequeued: {my_queue.dequeue()}")  # Expected: C
print(f"Front item (peek): {my_queue.get_front()}") # Expected: D

print(f"Dequeued: {my_queue.dequeue()}")  # Expected: D
print(f"Is queue empty now? {my_queue.is_empty()}") # Expected: True
print(f"Attempting to dequeue from empty queue: {my_queue.dequeue()}") # Expected: 'Queue is empty'
print(f"Attempting to peek from empty queue: {my_queue.get_front()}") # Expected: 'Queue is empty'
```

A Double-Ended Queue (Deque) is a linear data structure that allows you to insert and remove elements from both the front and the back with equal efficiency. It serves as a generalized, hybrid version of both a Stack (LIFO) and a regular Queue (FIFO).In Python, the primary conceptual and structural differences between a standard queue and a double-ended queue depend on whether you are looking at the pure data structure behavior or the standard libraries provided (queue.Queue vs collections.deque).

Queue applications - Task ordering, message buffering
Deque Applications - Sliding windows, undo-redo histories, palindromes|

```python
queue = []

# Enqueue (Adding to the end)
queue.append("Task 1")
queue.append("Task 2")
queue.append("Task 3")

print("Queue after enqueues:", queue)

# Dequeue (Removing from the front)
first_out = queue.pop(0)
print(f"Dequeued: {first_out}")
print("Queue after dequeue:", queue)
```

```python
from collections import deque

# Initialize deque
dq = deque([2, 3])

# Operations on the right end
dq.append(4)       # deque([2, 3, 4])
print(dq)

dq.pop()           # Removes 4 -> deque([2, 3])
# Operations on the left end
dq.appendleft(1)   # deque([1, 2, 3])
dq.popleft()       # Removes 1 -> deque([2, 3])
```

Let's break down your understanding for the CustomStack (using q1 and q2):

push(item): You're correct! When you push an item, it's simply added to q1 (our main_queue). At this point, q1 acts like a regular queue where the first item pushed is at the front, and the latest pushed is at the back. For example, if you push 10, then 20, then 30, q1 looks like [10, 20, 30].

pop() or peek() and the 'reversal': This is where the magic happens to simulate LIFO. To get the last item that was pushed (which is 30 in our example), we need to extract it from q1. But since q1 is a queue, its pop(0) operation would give us 10.

So, what we do is:

Move all but the last item from q1 to q2. In our example, 10 and 20 are moved to q2, so q2 becomes [10, 20] and q1 is [30].
The single remaining item in q1 (which is 30) is the top of our stack. We pop it from q1 (for the pop operation) or peek at it (q1[0]) (for the peek operation).
After we've gotten the 'top' item, we need to restore the other items. We effectively swap q1 and q2. The q2 (which now holds [10, 20]) becomes the new q1, and the old q1 (now empty) becomes the new q2. This prepares q1 for the next push operation, keeping the LIFO order for when we need to pop or peek again.
So, your intuition is spot on: the process involves reordering, but specifically, it's about isolating the last-in element by moving all other elements out of the way, and then putting them back in order. This achieves the LIFO behavior using FIFO queues. Well done!

Pushing Items: After pushing 10, 20, and 30, the stack is no longer empty. peek() correctly shows 30 as the top item, demonstrating LIFO.
Popping Items: Each pop() call removes the item that was most recently added. 30 is popped first, then 20.
Mixed Operations: When 40 is pushed, it becomes the new top item, and peek() reflects this. It's then popped before 20 (which was pushed earlier).
Empty Stack Handling: Finally, after all items are popped, is_empty() returns True, and attempting to pop() or peek() from the empty stack correctly returns "Stack is empty".
This confirms that the two-queue approach successfully provides LIFO behavior, even though it's built on FIFO queues!

```python
#2.Implement Stack using Queues
# Logic for using two queues to implement a stack:
# A stack is Last-In, First-Out (LIFO). A queue is First-In, First-Out (FIFO).
# To achieve LIFO behavior with two FIFO queues:
# 1. 'main_queue' (q1): This queue primarily holds the stack elements. When a new item is pushed, it's added here.
#    However, to simulate LIFO for 'pop' and 'peek', we need to ensure the *most recently added* item is readily accessible.
# 2. 'temp_queue' (q2): This is a helper queue used during 'pop' and 'peek' operations.
#
# When pushing an item:
#    The item is simply added to the 'main_queue'.
#
# When popping or peeking an item:
#    All elements except the very last one (which is the most recently added, and thus the 'top' of the stack) are moved from 'main_queue' to 'temp_queue'.
#    The remaining element in 'main_queue' is the top element. This element is then removed (for pop) or just read (for peek).
#    Finally, all elements are moved back from 'temp_queue' to 'main_queue' (or the queues are simply swapped).
#    This process ensures that the 'main_queue' always has its elements ordered such that the 'stack top' is the last element to be removed from the queue.
class CustomStack:
    def __init__(self):
        # Initialize two queues using Python lists. We'll treat q1 as the primary queue.
        self.q1 = []  # Main queue, will store elements (eventually with the top at the end)
        self.q2 = []  # Helper queue, used for reordering during pop/peek

    def push(self, item):
        """Adds an item to the top of the stack."""
        # Simply add the new item to the main queue (q1).
        # The reordering to simulate LIFO will happen during pop/peek.
        self.q1.append(item) # Enqueue the item into q1

    def pop(self):
        """Removes and returns the item from the top of the stack."""
        # If the stack is empty, return an appropriate message.
        if self.is_empty():
            return "Stack is empty"

        # Move all elements from q1 to q2 except the last one.
        # The last element in q1 is the 'top' of our stack (the most recently pushed item).
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0)) # Dequeue from q1 and enqueue to q2

        # The single remaining element in q1 is the item we want to pop.
        popped_item = self.q1.pop(0) # Dequeue the top element from q1

        # Swap the names of q1 and q2. Now q2 (which contains all the other elements)
        # becomes the new main queue, and q1 is empty for future use.
        self.q1, self.q2 = self.q2, self.q1

        return popped_item

    def peek(self):
        """Returns the item at the top of the stack without removing it."""
        # If the stack is empty, return an appropriate message.
        if self.is_empty():
            return "Stack is empty"

        # Move all elements from q1 to q2 except the last one (the 'top' of the stack).
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0)) # Dequeue from q1 and enqueue to q2

        # The single remaining element in q1 is the item we want to peek.
        peeked_item = self.q1[0] # Peek at the top element without removing it

        # Now, move this peeked item from q1 to q2 as well.
        # All original elements are now in q2, in their original relative order.
        self.q2.append(self.q1.pop(0))

        # Swap the names of q1 and q2. Now q2 (containing all elements) becomes the new main queue.
        self.q1, self.q2 = self.q2, self.q1

        return peeked_item

    def is_empty(self):
        """Checks if the stack has any elements."""
        # The stack is empty if its main queue (q1) is empty.
        return len(self.q1) == 0

# --- Testing the Implementation ---:
print("\n--- Testing CustomStack Implementation ---")
my_stack = CustomStack()
print(f"Is stack empty initially? {my_stack.is_empty()}") # Expected: True

my_stack.push(10)
my_stack.push(20)
my_stack.push(30)
print(f"Stack after pushing 10, 20, 30. Is empty? {my_stack.is_empty()}") # Expected: False
print(f"Top item (peek): {my_stack.peek()}") # Expected: 30

print(f"Popped: {my_stack.pop()}")  # Expected: 30 (Last In, First Out)
print(f"Top item (peek): {my_stack.peek()}") # Expected: 20

my_stack.push(40)
print(f"Stack after pushing 40. Top item: {my_stack.peek()}") # Expected: 40

print(f"Popped: {my_stack.pop()}")  # Expected: 40
print(f"Popped: {my_stack.pop()}")  # Expected: 20
print(f"Top item (peek): {my_stack.peek()}") # Expected: 10

print(f"Popped: {my_stack.pop()}")  # Expected: 10
print(f"Is stack empty now? {my_stack.is_empty()}") # Expected: True
print(f"Attempting to pop from empty stack: {my_stack.pop()}") # Expected: 'Stack is empty'
print(f"Attempting to peek from empty stack: {my_stack.peek()}") # Expected: 'Stack is empty'
```
