```python
def search_iterative(nums, target):
    # Establish boundary markers at the extreme ends of the array
    low = 0
    high = len(nums) - 1

    # Keep searching as long as our boundaries haven't crossed past each other
    while low <= high:
        # Calculate the exact middle index of our current search window
        mid = (low + high) // 2

        # Scenario 1: The middle element is our target
        if nums[mid] == target:
            return mid

        # Scenario 2: The middle element is smaller than our target.
        # This means the target must be on the right side of 'mid'.
        elif nums[mid] < target:
            low = mid + 1

        # Scenario 3: The middle element is larger than our target.
        # This means the target must be on the left side of 'mid'.
        else:
            high = mid - 1

    # If the boundaries cross and the loop exits, the target does not exist
    return -1
print(search_iterative([1,2,3,4,5,6,7], 5))
```

```python
def search_recursive_helper(nums, target, low, high):
    # Base Case 1: If low crosses high, the element is not in the array
    if low > high:
        return -1

    # Calculate middle index
    mid = (low + high) // 2

    # Base Case 2: Element found at the middle index
    if nums[mid] == target:
        return mid

    # Recursive Case 1: Middle value is smaller than target.
    # Search the right half by passing a modified 'low' boundary.
    elif nums[mid] < target:
        return search_recursive_helper(nums, target, mid + 1, high)

    # Recursive Case 2: Middle value is larger than target.
    # Search the left half by passing a modified 'high' boundary.
    else:
        return search_recursive_helper(nums, target, low, mid - 1)


# Main function called by the user
def search_recursive(nums, target):
    # Kick off the recursion by passing the initial boundary pointers
    return search_recursive_helper(nums, target, 0, len(nums) - 1)

print(search_recursive([1,2,3,4,5,6,7], 5))
```

```python
#1. Subsets using Backtracking

def subsets_backtracking(nums):
    result = [] # This will store all the generated subsets
    current_subset = [] # This will store the subset being built in the current recursive call

    def backtrack(start_index):
        print(f"\n--- Entering backtrack(start_index={start_index}) ---")
        print(f"  Before adding: current_subset={list(current_subset)}, result={result}")

        # Base case / Decision point: Add the current_subset to the result list
        # A 'subset' is formed at every point of the recursion tree, including the empty set.
        result.append(list(current_subset)) # Append a copy, not the reference
        print(f"  Added subset: {list(current_subset)}. Result now: {result}")

        # Explore choices for the current level of recursion
        # We iterate from `start_index` to avoid duplicate subsets and maintain order
        for i in range(start_index, len(nums)):
            # 1. Choose: Include the current element nums[i] in the subset
            current_subset.append(nums[i])
            print(f"  CHOICE: Including {nums[i]}. current_subset={list(current_subset)}")

            # 2. Explore: Recurse to find subsets with this chosen element
            # We pass i + 1 to ensure we only consider elements after the current one
            backtrack(i + 1)

            # 3. Unchoose (Backtrack): Remove the element to explore other possibilities
            # This is the backtracking step, undoing the previous choice.
            removed_element = current_subset.pop()
            print(f"  UNCHOICE (Backtracking): Removing {removed_element}. current_subset={list(current_subset)}")

        print(f"--- Exiting backtrack(start_index={start_index}) ---")

    # Start the backtracking process from the beginning of the array (index 0)
    backtrack(0)
    return result

# Example usage with a small set:
nums_example = [1, 2]
print(f"\n>>> Generating all subsets for {nums_example} using backtracking: <<<")
all_subsets = subsets_backtracking(nums_example)
print(f"\nFinal list of all subsets for {nums_example}: {all_subsets}")
```

```python
#2. Permutations using Backtracking

def permutations_backtracking_explained(nums):
    result = [] # Stores all complete permutations
    current_permutation = [] # Stores the permutation being built
    used = [False] * len(nums) # Tracks which elements have been used

    def backtrack():
        print(f"\n--- Entering backtrack() ---")
        print(f"  Current state: current_permutation={list(current_permutation)}, used={used}, result={result}")

        # Base Case: If the current permutation is complete (same length as original nums)
        if len(current_permutation) == len(nums):
            result.append(list(current_permutation)) # Add a copy to the result
            print(f"  BASE CASE: Permutation complete. Added {list(current_permutation)}. Result now: {result}")
            return

        # Explore choices for the current level
        for i in range(len(nums)):
            # If the number at index `i` has already been used, skip it
            if used[i]:
                print(f"  SKIP: nums[{i}]={nums[i]} already used.")
                continue

            # 1. Choose: Include the current element nums[i]
            current_permutation.append(nums[i])
            used[i] = True # Mark it as used
            print(f"  CHOICE: Including {nums[i]}. current_permutation={list(current_permutation)}, used={used}")

            # 2. Explore: Recurse to find permutations with this chosen element
            backtrack()

            # 3. Unchoose (Backtrack): Remove the element and mark it as unused
            # This allows other paths to use this element.
            removed_element = current_permutation.pop()
            used[i] = False
            print(f"  UNCHOICE (Backtracking): Removing {removed_element}. current_permutation={list(current_permutation)}, used={used}")

        print(f"--- Exiting backtrack() ---")

    # Start the backtracking process
    backtrack()
    return result

# Example usage with a small set:
nums_example = [1, 2, 3]
print(f"\n>>> Generating all permutations for {nums_example} using backtracking: <<<")
all_permutations = permutations_backtracking_explained(nums_example)
print(f"\nFinal list of all permutations for {nums_example}: {all_permutations}")
```
