```python
def count_ways_to_climb(total_steps):
    # Base Cases: If the staircase has 1 or 2 steps,
    # we already know the answer instantly.
    if total_steps == 1:
        return 1
    if total_steps == 2:
        return 2

    # Set up our initial records for the first two steps
    ways_for_step_1 = 1  # There is 1 way to reach step 1
    ways_for_step_2 = 2  # There are 2 ways to reach step 2

    # Start calculating from Step 3 up to our target total_steps
    for current_step in range(3, total_steps + 1):

        # Rule: To get to the current step, we add the distinct ways
        # of the two steps directly behind it.
        ways_for_current_step = ways_for_step_2 + ways_for_step_1

        # --- THE SHIFT ---
        # Slide our records forward to prepare for the next step loop:
        # Step 1's value is no longer needed; it gets replaced by Step 2's value.
        ways_for_step_1 = ways_for_step_2

        # Step 2's value gets replaced by the current step's value we just found.
        ways_for_step_2 = ways_for_current_step

    # Once the loop finishes, ways_for_step_2 will hold the final total
    return ways_for_step_2


# --- Test Cases to verify ---

# Test 1: 3 steps should give 3 ways
print(f"Ways to climb 3 steps: {count_ways_to_climb(3)}")

# Test 2: 4 steps should give 5 ways
print(f"Ways to climb 4 steps: {count_ways_to_climb(4)}")

# Test 3: 5 steps should give 8 ways
print(f"Ways to climb 5 steps: {count_ways_to_climb(5)}")
```

```python
def find_max_profit(prices):
    if not prices:
        return 0

    max_profit = 0
    # Start by assuming the first day is the lowest price
    lowest_price_so_far = prices[0]

    for current_price in prices:
        # 1. Update the lowest price if we found a cheaper day to buy
        if current_price < lowest_price_so_far:
            lowest_price_so_far = current_price

        # 2. Calculate potential profit if we sold today
        potential_profit = current_price - lowest_price_so_far

        # 3. Keep the best profit seen so far
        if potential_profit > max_profit:
            max_profit = potential_profit

    return max_profit

# Example usage:
stock_prices = [7, 1, 5, 3, 6, 4]
# Buy at 1, Sell at 6 -> Profit = 5
print(find_max_profit(stock_prices))  # Output: 5
```

PRIORITY QUEUE

```python
import heapq
# Initialize the priority queue
priority_queue = []
# Insert elements: (Priority_Number, Task_Name)
# Note: Lower number = Higher priority in a Min-Heap
heapq.heappush(priority_queue, (3, "Clear emails"))
heapq.heappush(priority_queue, (1, "Fix production crash"))
heapq.heappush(priority_queue, (2, "Prepare for meeting"))
# Process tasks based on priority
while priority_queue:
    priority, task = heapq.heappop(priority_queue)
    print(task,priority)
```

Finding kth largest element

```python
import heapq

def find_kth_largest(nums, k):
    # Initialize an empty min-heap
    min_heap = []

    for num in nums:
        # Push the current number onto the heap
        heapq.heappush(min_heap, num)

        # If the heap grows larger than k, remove the smallest element
        if len(min_heap) > k:
            heapq.heappop(min_heap)

    # The root of the heap is now the kth largest element
    return min_heap[0]

# --- Example Usage ---
numbers = [3, 2, 1, 5, 6, 4]
k_value = 2

result = find_kth_largest(numbers, k_value)
print(f"The {k_value}nd largest element is: {result}")
# Output: The 2nd largest element is: 5
```

max heap

```python
import heapq

max_heap = []
numbers = [10, 4, 15, 1]

# Invert values to simulate a Max-Heap
for num in numbers:
    heapq.heappush(max_heap, -num)

# Popping will now return the largest original number
largest = -heapq.heappop(max_heap)
print(f"Largest element: {largest}") # Output: 15
```
