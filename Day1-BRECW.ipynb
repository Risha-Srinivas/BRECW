```python
#array - running sum
nums = [1,2,3,4]

result = []

for i in range(len(nums)):

    total = 0

    for j in range(i + 1):#0,1,2,3
        total += nums[j]

    result.append(total)

print(result)
```

```python
#ruuning sum optimized
nums = [1,2,3,4]

result = []

for i in range(len(nums)):

    total = 0

    for j in range(i + 1):
        total += nums[j]

    result.append(total)
    print(result)

print('final ',result)
```

```python
#pivot table - find point where left sum == right sum

nums = [1,7,3,6,5,6]

for i in range(len(nums)):

    left_sum = sum(nums[:i])

    right_sum = sum(nums[i+1:])

    if left_sum == right_sum:
        print(i,'found')
        break
```

```python
#optimized
nums = [1,7,3,6,5,6]

total = sum(nums)

left_sum = 0

for i in range(len(nums)):

    print('loop of i ', i, 'left_sum is ', left_sum)

    right_sum = total - left_sum - nums[i]
    print('right sum is ',right_sum)
    if left_sum == right_sum:
        print(i)
        break

    left_sum += nums[i]
```

```python
#pattern - moving left and right pointers in opp direction - update each index based on condition
s = "madam"

left = 0
right = len(s) - 1

while left < right:

    if s[left] != s[right]:
        print(False)
        break

    left += 1
    right -= 1

else:
    print(True)
```

```python
#pattern - same moving pointers in opp direction but swap instead of check
s_s = input("Enter a string: ")
s = list(s_s)

left = 0
right = len(s) - 1

while left < right:

    s[left], s[right] = s[right], s[left]

    left += 1
    right -= 1

print(s)
```

```python
#3. Move Zeroes - move pointers in same direction

nums = [0,0,5,1,2,6]
slow = 0
for fast in range(len(nums)):
  if nums[fast] != 0 and nums[slow] == 0:
    nums[slow], nums[fast] = nums[fast], nums[slow]
    print('nums ',nums)
    print('fast ',fast)
    if nums[slow] != 0:
      slow += 1
      print('slow ',slow)
print(nums)
```

```python
#4.sort square of ints - opp pointers - compare
nums = [-4,-1,0,3,10]
result = [0] * len(nums)
left = 0
right = len(nums) - 1
position = len(nums) - 1
print(nums)
while left <= right:
    if abs(nums[left]) > abs(nums[right]):
        result[position] = nums[left] ** 2
        left += 1
        print(result)
    else:
        result[position] = nums[right] ** 2
        right -= 1
        print(result)
    position -= 1
print(result)
```
