## Sliding Window

### Problem 1: Maximum Average Subarray

**Problem Statement:**
Given an array consisting of `n` integers and a positive integer `k`, find the contiguous subarray whose length is equal to `k` that has the maximum average value. You need to output the maximum average value.

**Input/Output Example:**

**Input:**
```
nums = [1, 12, -5, -6, 50, 3], k = 4
```

**Output:**
```
12.75
```

**Explanation:**
The contiguous subarray of length 4 with the maximum average value is `[12, -5, -6, 50]`. Its sum is `12 + (-5) + (-6) + 50 = 51`. The average is `51 / 4 = 12.75`.

**Thought Process:**

1.  **Brute-Force Approach:** A straightforward way would be to consider every possible subarray of length `k`. For each subarray, calculate its sum, then its average, and keep track of the maximum average found. This would involve nested loops: an outer loop for the starting position of the subarray and an inner loop to sum elements up to `k` positions. This approach would have a time complexity of `O(n*k)`, which might be inefficient for large arrays.

2.  **Optimizing with Sliding Window:** Can we improve upon the brute-force method? Notice that when we move from one subarray `[nums[i], ..., nums[i+k-1]]` to the next `[nums[i+1], ..., nums[i+k]]`, most elements in the sum remain the same. We only subtract the element `nums[i]` that leaves the window and add `nums[i+k]` that enters the window. This is the core idea of the *sliding window* technique.

3.  **Algorithm Idea:**
    *   First, calculate the sum of the initial window (the first `k` elements). This sum will be our `current_sum` and also our initial `max_sum` (since it's the first window we've considered).
    *   Then, iterate through the rest of the array, starting from the `k`-th element. In each step:
        *   Update `current_sum` by subtracting the element that is no longer in the window (which is `nums[i-k]`) and adding the new element that just entered the window (`nums[i]`).
        *   Compare this `current_sum` with `max_sum` and update `max_sum` if `current_sum` is greater.
    *   After iterating through all possible windows, `max_sum` will hold the sum of the subarray with the maximum average. Divide `max_sum` by `k` to get the maximum average value.

**Algorithm:**

1.  Initialize `current_sum` to 0.
2.  Calculate the sum of the first `k` elements and store it in `current_sum`.
3.  Initialize `max_sum` with the value of `current_sum`.
4.  Iterate from `i = k` to `len(nums) - 1`:
    a.  Update `current_sum` by subtracting `nums[i - k]` (element leaving the window) and adding `nums[i]` (element entering the window).
    b.  Update `max_sum = max(max_sum, current_sum)`.
5.  Return `max_sum / k`.

````python
def findMaxAverage(nums, k):
    # Calculate the sum of the first window (first k elements)
    current_sum = sum(nums[:k])
    max_sum = current_sum

    # Slide the window from the k-th element to the end
    for i in range(k, len(nums)):
        # Subtract the element leaving the window and add the element entering the window
        current_sum = current_sum - nums[i - k] + nums[i]
        # Update max_sum if the current window's sum is greater
        max_sum = max(max_sum, current_sum)

    # The maximum average is the max_sum divided by k
    return max_sum / k
print(findMaxAverage([1,2,3,4,5,6,7,8],2))
# Test cases
print(f"Maximum average for [1,12,-5,-6,50,3], k=4: {findMaxAverage([1,12,-5,-6,50,3], 4)}") # Expected: 12.75
print(f"Maximum average for [0,1,1,1,1], k=2: {findMaxAverage([0,1,1,1,1], 2)}")   # Expected: 1.0
print(f"Maximum average for [5], k=1: {findMaxAverage([5], 1)}")             # Expected: 5.0
print(f"Maximum average for [1,12,-5,-6,50,3], k=1: {findMaxAverage([1,12,-5,-6,50,3], 1)}") # Expected: 50.0
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   The `sum(nums[:k])` operation takes `O(k)` time to calculate the sum of the first window.
    *   The subsequent `for` loop iterates `N - k` times. Inside the loop, each operation (subtraction, addition, comparison) takes `O(1)` time.
    *   Therefore, the total time complexity is `O(k + (N - k))`, which simplifies to `O(N)`.

*   **Space Complexity:** `O(1)`
    *   We only use a few constant extra variables (`current_sum`, `max_sum`, `i`), regardless of the input array size.

### Problem 2: Max Consecutive Ones

**Problem Statement:**
Given a binary array `nums`, return the maximum number of consecutive `1`s in the array.

**Input/Output Example:**

**Input:**
```
nums = [1,1,0,1,1,1]
```

**Output:**
```
3
```

**Explanation:**
The first two digits are `1`s (`[1,1,0,1,1,1]`). The last three digits are `1`s (`[1,1,0,1,1,1]`). The maximum number of consecutive `1`s is 3.

**Thought Process:**

1.  **Simple Iteration:** This problem can be solved by iterating through the array once. We need to keep track of the current count of consecutive `1`s and the maximum count found so far.

2.  **How to track:**
    *   When we encounter a `1`, we increment our `current_count`.
    *   When we encounter a `0`, the sequence of `1`s is broken. So, we compare our `current_count` with `max_count` and update `max_count` if `current_count` is larger. Then, we reset `current_count` to `0` because a new sequence of `1`s (if any) will start.
    *   It's crucial to also perform the `max_count` update one last time *after* the loop finishes. This handles the case where the array ends with a sequence of `1`s (e.g., `[1,1,1]`), as the `0` condition to reset and update `max_count` wouldn't be triggered for the last sequence.

**Algorithm:**

1.  Initialize `max_consecutive_ones` to 0.
2.  Initialize `current_consecutive_ones` to 0.
3.  Iterate through each `num` in the `nums` array:
    a.  If `num` is `1`, increment `current_consecutive_ones`.
    b.  If `num` is `0`:
        i.  Update `max_consecutive_ones = max(max_consecutive_ones, current_consecutive_ones)`.
        ii. Reset `current_consecutive_ones` to 0.
4.  After the loop finishes, perform one final update: `max_consecutive_ones = max(max_consecutive_ones, current_consecutive_ones)`. This step is important for arrays ending with `1`s.
5.  Return `max_consecutive_ones`.

````python
def findMaxConsecutiveOnes(nums):
    max_consecutive_ones = 0
    current_consecutive_ones = 0
    for num in nums:
        if num == 1:
            current_consecutive_ones += 1
        else:
            # If we see a 0, the current sequence of 1s is broken.
            # Update max_consecutive_ones and reset current_consecutive_ones.
            max_consecutive_ones = max(max_consecutive_ones, current_consecutive_ones)
            current_consecutive_ones = 0
    # After the loop, one final check is needed in case the array ends with 1s.
    max_consecutive_ones = max(max_consecutive_ones, current_consecutive_ones)
    return max_consecutive_ones

# Test cases
print(f"Max consecutive ones for [1,1,0,1,1,1]: {findMaxConsecutiveOnes([1,1,0,1,1,1])}") # Expected: 3
print(f"Max consecutive ones for [1,0,1,1,0,1]: {findMaxConsecutiveOnes([1,0,1,1,0,1])}") # Expected: 2
print(f"Max consecutive ones for [0,0,0,0]: {findMaxConsecutiveOnes([0,0,0,0])}")   # Expected: 0
print(f"Max consecutive ones for [1,1,1,1]: {findMaxConsecutiveOnes([1,1,1,1])}")   # Expected: 4
print(f"Max consecutive ones for [0,1,1,1,0,1,1]: {findMaxConsecutiveOnes([0,1,1,1,0,1,1])}") # Expected: 3
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   We iterate through the array `nums` exactly once. Each operation inside the loop (comparison, increment, `max` function) takes constant time.
    *   Therefore, the total time complexity is directly proportional to the number of elements in the array, `N`.

*   **Space Complexity:** `O(1)`
    *   We only use a few constant extra variables (`max_consecutive_ones`, `current_consecutive_ones`), regardless of the input array size.

## Strings

### Problem 1: Valid Anagram

**Problem Statement:**
Given two strings `s` and `t`, return `true` if `t` is an anagram of `s`, and `false` otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Input/Output Example:**

**Example 1:**
**Input:**
```
s = "anagram", t = "nagaram"
```
**Output:**
```
true
```

**Example 2:**
**Input:**
```
s = "rat", t = "car"
```
**Output:**
```
false
```

**Thought Process:**

1.  **What defines an anagram?** Two strings are anagrams of each other if they contain the same characters with the same frequencies. For example, 'listen' and 'silent' both have one 'l', one 'i', one 's', one 't', one 'e', and one 'n'.

2.  **Initial Checks:**
    *   If the lengths of the two strings `s` and `t` are different, they cannot be anagrams. So, this is a quick early exit condition.

3.  **Approach 1: Sorting:**
    *   If two strings are anagrams, sorting their characters alphabetically should result in identical strings. For example, sorting "anagram" gives "aaagmnr", and sorting "nagaram" also gives "aaagmnr".
    *   This is a simple and effective approach.

4.  **Approach 2: Character Counting (Hash Map/Frequency Array):**
    *   We can count the frequency of each character in both strings.
    *   If `s` and `t` are anagrams, then for every character from 'a' to 'z', its count in `s` must be equal to its count in `t`.
    *   We can use a hash map (dictionary in Python) or a frequency array (since we are dealing with lowercase English letters, an array of size 26 is sufficient).
    *   **Algorithm using a hash map:**
        *   Create a hash map (e.g., `s_counts`) to store character frequencies for string `s`.
        *   Iterate through `s`, incrementing the count for each character in `s_counts`.
        *   Create another hash map (e.g., `t_counts`) for string `t` and do the same.
        *   Compare `s_counts` and `t_counts`. If they are identical, `t` is an anagram of `s`.
    *   **Optimization for hash map:** Instead of two separate hash maps, we can use one. Increment counts for characters in `s` and decrement counts for characters in `t`. If `s` and `t` are anagrams, all counts in the map should be zero at the end.

5.  **Choosing the best approach:**
    *   Sorting is generally `O(N log N)` where `N` is the length of the string. While simple, it might be slower for very long strings.
    *   Character counting using a hash map is `O(N)` because we iterate through the strings once or twice. This is generally more efficient for larger inputs.

**Algorithm (using Character Counting - optimized):**

1.  **Length Check:** If `len(s)` is not equal to `len(t)`, return `false` immediately.
2.  **Frequency Map:** Create a dictionary, `char_counts`, to store character frequencies.
3.  **Process String `s`:** Iterate through each character `char` in `s`:
    a.  Increment `char_counts[char]`.
4.  **Process String `t`:** Iterate through each character `char` in `t`:
    a.  Decrement `char_counts[char]`.
5.  **Verify Frequencies:** Iterate through all values in `char_counts`:
    a.  If any value is not `0`, it means the character frequencies do not match, so return `false`.
6.  If all checks pass, return `true`.

````python
def isAnagram(s, t):
    # If lengths are different, they cannot be anagrams
    if len(s) != len(t):
        return False
    # Use a standard dictionary to store character counts.
    char_counts = {}
    # Count character frequencies in string s
    for char in s:
        char_counts[char] = char_counts.get(char, 0) + 1
    # Decrement character frequencies for string t
    for char in t:
        # If character from t is not in s's counts, it's not an anagram
        if char not in char_counts:
            return False
        # If char from t is in s's counts, but its count is already exhausted or negative
        if char_counts[char] <= 0:
            return False
        char_counts[char] -= 1

    # After processing both strings, all counts in char_counts should be zero
    # if they are anagrams. Since we checked `char_counts[char] <= 0` in the second loop,
    # and `len(s) == len(t)`, if we reach here, all counts must implicitly be zero.
    # So, an explicit loop through char_counts is not strictly needed for correctness.

    return True

# Test cases
print(f"'anagram' and 'nagaram' are anagrams: {isAnagram('anagram', 'nagaram')}") # Expected: True
print(f"'rat' and 'car' are anagrams: {isAnagram('rat', 'car')}")             # Expected: False
print(f"'listen' and 'silent' are anagrams: {isAnagram('listen', 'silent')}")   # Expected: True
print(f"'hello' and 'world' are anagrams: {isAnagram('hello', 'world')}")     # Expected: False
print(f"'a' and 'ab' are anagrams: {isAnagram('a', 'ab')}")                 # Expected: False
print(f"'aabb' and 'bbaa' are anagrams: {isAnagram('aabb', 'bbaa')}")         # Expected: True
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   `len(s) != len(t)` check is `O(1)`.
    *   Iterating through string `s` to populate `char_counts` takes `O(N)` time, where `N` is the length of `s` (and `t`).
    *   Iterating through string `t` to decrement `char_counts` takes `O(N)` time.
    *   Dictionary operations (insert, increment, decrement) on average take `O(1)` time.
    *   Therefore, the total time complexity is `O(N) + O(N) = O(N)`.

*   **Space Complexity:** `O(1)`
    *   The `char_counts` dictionary will store at most 26 entries (for lowercase English letters).
    *   Since the number of distinct characters is constant (alphabet size), the space complexity is considered `O(1)` regardless of the input string length.
    *   If the input could contain extended Unicode characters, the space complexity would be `O(C)` where `C` is the number of unique characters in the input strings.

### Problem 2: First Unique Character in a String

**Problem Statement:**
Given a string `s`, find the first non-repeating character in it and return its index. If it does not exist, return -1.

**Input/Output Example:**

**Example 1:**
**Input:**
```
s = "leetcode"
```
**Output:**
```
0
```
**Explanation:** 'l' is the first unique character, and its index is 0.

**Example 2:**
**Input:**
```
s = "loveleetcode"
```
**Output:**
```
2
```
**Explanation:** 'l' appears twice. 'o' appears twice. 'v' is the first unique character, and its index is 2.

**Example 3:**
**Input:**
```
s = "aabb"
```
**Output:**
```
-1
```
**Explanation:** All characters appear twice, so there is no unique character.

**Thought Process:**

1.  **What does "first non-repeating character" mean?** It means the character that appears exactly once in the string, and among all such characters, it's the one that appears earliest in the string.

2.  **How to track character occurrences?** We need to know how many times each character appears in the string. A hash map (dictionary in Python) is ideal for this, where keys are characters and values are their counts.

3.  **Step-by-step approach:**
    *   **First Pass (Counting):** Iterate through the string once to build the frequency map for all characters. This will tell us which characters are unique (count = 1) and which are repeating (count > 1).
    *   **Second Pass (Finding First Unique):** After we have the counts, iterate through the string *again*. For each character, check its count in our frequency map. The first character we encounter that has a count of `1` is our answer. Its index will be the current index from the second pass.
    *   **No Unique Character:** If we finish the second pass without finding any character with a count of `1`, it means no unique character exists, so we return -1.

**Algorithm:**

1.  Create a dictionary `char_counts` to store character frequencies.
2.  **First Pass:** Iterate through the input string `s` from left to right:
    a.  For each character, increment its count in `char_counts`.
3.  **Second Pass:** Iterate through the input string `s` *again* from left to right, also keeping track of the current index `i`:
    a.  For the character at the current index `s[i]`:
        i.  Check its count in `char_counts[s[i]]`.
        ii. If `char_counts[s[i]]` is `1`, then this is the first non-repeating character. Return its index `i`.
4.  If the loop completes without returning, it means no unique character was found. Return `-1`.

````python
from collections import defaultdict

def firstUniqChar(s: str) -> int:
    # Step 1: Create a frequency map for all characters
    char_counts = defaultdict(int)
    for char in s:
        char_counts[char] += 1
    # Step 2: Iterate through the string again to find the first character with count 1
    for i in range(len(s)):
        if char_counts[s[i]] == 1:
            return i
    # If no unique character is found after checking all characters
    return -1
# Test cases
print(f"First unique character index for 'leetcode': {firstUniqChar('leetcode')}")     # Expected: 0
print(f"First unique character index for 'loveleetcode': {firstUniqChar('loveleetcode')}") # Expected: 2
print(f"First unique character index for 'aabb': {firstUniqChar('aabb')}")         # Expected: -1
print(f"First unique character index for 'a': {firstUniqChar('a')}")             # Expected: 0
print(f"First unique character index for '': {firstUniqChar('')}")               # Expected: -1
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   The first pass to populate the `char_counts` dictionary takes `O(N)` time, where `N` is the length of the string `s`.
    *   The second pass to find the first unique character also takes `O(N)` time in the worst case (e.g., if the unique character is at the end, or if no unique character exists).
    *   Dictionary operations (insert, lookup) take `O(1)` time on average.
    *   Therefore, the total time complexity is `O(N) + O(N) = O(N)`.

*   **Space Complexity:** `O(1)`
    *   The `char_counts` dictionary will store at most 26 entries (for lowercase English letters).
    *   Since the number of distinct characters is constant (alphabet size), the space complexity is considered `O(1)` regardless of the input string length.
    *   Similar to the Anagram problem, if the input could contain extended Unicode characters, the space complexity would be `O(C)` where `C` is the number of unique characters in the input string.

### Problem 3: Isomorphic Strings

**Problem Statement:**
Given two strings `s` and `t`, determine if they are isomorphic.

Two strings `s` and `t` are isomorphic if the characters in `s` can be replaced to get `t`.

All occurrences of a character must be replaced with another character while preserving the order of the characters. No two characters may map to the same character, but a character may map to itself.

**Input/Output Example:**

**Example 1:**
**Input:**
```
s = "egg", t = "add"
```
**Output:**
```
true
```
**Explanation:** 'e' maps to 'a', 'g' maps to 'd'. The mapping is consistent.

**Example 2:**
**Input:**
```
s = "foo", t = "bar"
```
**Output:**
```
false
```
**Explanation:** 'f' maps to 'b', 'o' maps to 'a'. But the second 'o' in 'foo' should map to 'a' as well, but 'r' is different.

**Example 3:**
**Input:**
```
s = "paper", t = "title"
```
**Output:**
```
true
```
**Explanation:** 'p' maps to 't', 'a' maps to 'i', 'e' maps to 'l', 'r' maps to 'e'. All consistent.

**Thought Process:**

1.  **Understanding Isomorphism:** The key idea is a one-to-one mapping. Each character in `s` must map to exactly one character in `t`, and no two distinct characters in `s` can map to the same character in `t`. This mapping must also be consistent throughout the strings.

2.  **Initial Check:** If the lengths of `s` and `t` are different, they cannot be isomorphic. This is a quick early exit.

3.  **How to track mappings?** We need to store the mapping from `s` to `t` and also ensure that `t` characters are not mapped to by multiple `s` characters. Two hash maps (dictionaries) can help:
    *   `s_to_t_map`: Stores `s_char -> t_char` mappings.
    *   `t_to_s_map`: Stores `t_char -> s_char` mappings (or simply, tracks which characters in `t` have already been mapped to).

4.  **Iterating and Building Mappings:**
    *   We'll iterate through both strings simultaneously, character by character at the same index `i`.
    *   Let `char_s = s[i]` and `char_t = t[i]`.
    *   **Case 1: `char_s` is already in `s_to_t_map`:**
        *   This means `char_s` has a pre-established mapping. We must check if its current mapping (`s_to_t_map[char_s]`) is consistent with `char_t`. If `s_to_t_map[char_s] != char_t`, then the mapping is broken, and they are not isomorphic. Return `false`.
    *   **Case 2: `char_s` is NOT in `s_to_t_map`:**
        *   This is a new character from `s`. We need to establish a new mapping.
        *   Before creating the mapping `char_s -> char_t`, we must ensure that `char_t` has not already been mapped to by a *different* character from `s`. This is where `t_to_s_map` (or a `seen_t_chars` set) comes in. If `char_t` is already in `t_to_s_map`, it means some other `s_char_prev` was mapped to `char_t`. Since `char_s` is different from `s_char_prev` (because `char_s` is new), this violates the one-to-one rule. Return `false`.
        *   If `char_t` has not been mapped to yet, then establish the new mappings: `s_to_t_map[char_s] = char_t` and add `char_t` to `t_to_s_map` (or `t_to_s_map[char_t] = char_s`).

5.  **Conclusion:** If the loop completes without returning `false`, it means all characters were mapped consistently, satisfying all conditions. Return `true`.

**Algorithm:**

1.  **Length Check:** If `len(s)` is not equal to `len(t)`, return `False`.
2.  Initialize two dictionaries:
    *   `map_s_to_t`: To store mappings from characters in `s` to characters in `t`.
    *   `map_t_to_s`: To store mappings from characters in `t` to characters in `s` (ensures no two characters from `s` map to the same character in `t`).
3.  Iterate from `i = 0` to `len(s) - 1`:
    a.  Let `char_s = s[i]` and `char_t = t[i]`.
    b.  **Check `map_s_to_t`:**
        i.  If `char_s` is already in `map_s_to_t`:
            *   If `map_s_to_t[char_s]` is not equal to `char_t`, return `False` (inconsistent mapping).
        ii. Else (`char_s` is not in `map_s_to_t`):
            *   **Check `map_t_to_s`:** If `char_t` is already in `map_t_to_s`:
                *   This means `char_t` has already been mapped by a *different* character from `s`. Return `False` (violates one-to-one rule).
            *   Else (`char_t` is not in `map_t_to_s`):
                *   Establish new mappings: `map_s_to_t[char_s] = char_t` and `map_t_to_s[char_t] = char_s`.
4.  If the loop finishes, all checks passed, so return `True`.

````python
def isIsomorphic(s: str, t: str) -> bool:
    map_S_to_T = {}
    map_T_to_S = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        # Check s -> t mapping consistency
        if char_s in map_S_to_T:
            if map_S_to_T[char_s] != char_t:
                return False
        else:
            map_S_to_T[char_s] = char_t

        # Check t -> s mapping consistency
        if char_t in map_T_to_S:
            if map_T_to_S[char_t] != char_s:
                return False
        else:
            map_T_to_S[char_t] = char_s

    return True
# Test cases
print(f"'egg' and 'add' are isomorphic: {isIsomorphic('egg', 'add')}")         # Expected: True
print(f"'foo' and 'bar' are isomorphic: {isIsomorphic('foo', 'bar')}")         # Expected: False
print(f"'paper' and 'title' are isomorphic: {isIsomorphic('paper', 'title')}") # Expected: True
print(f"'ab' and 'aa' are isomorphic: {isIsomorphic('ab', 'aa')}")           # Expected: False (a->a, b->a; two s-chars map to one t-char)
print(f"'ab' and 'ca' are isomorphic: {isIsomorphic('ab', 'ca')}")           # Expected: True
print(f"'badc' and 'baba' are isomorphic: {isIsomorphic('badc', 'baba')}")     # Expected: False (b->b, a->a, d->b; 'd' attempts to map to 'b' which is already mapped by 'b')
print(f"'s' and 's' are isomorphic: {isIsomorphic('s', 's')}")               # Expected: True
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   The initial length check is `O(1)`.
    *   We iterate through the strings once using a single `for` loop, where `N` is the length of the strings.
    *   Dictionary operations (insert, lookup) take `O(1)` time on average.
    *   Therefore, the total time complexity is `O(N)`.

*   **Space Complexity:** `O(1)`
    *   In the worst case, the `map_s_to_t` and `map_t_to_s` dictionaries will store mappings for all unique characters in the strings.
    *   For lowercase English letters, the maximum number of unique characters is 26.
    *   Since the number of distinct characters is constant (alphabet size), the space complexity is considered `O(1)` regardless of the input string length.
    *   If the input could contain extended Unicode characters, the space complexity would be `O(C)` where `C` is the number of unique characters in the input strings.

## Hash Map and Hash Set

### Problem 1: Two Sum

**Problem Statement:**
Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

**Input/Output Example:**

**Example 1:**
**Input:**
```
nums = [2, 7, 11, 15], target = 9
```
**Output:**
```
[0, 1]
```
**Explanation:** Because `nums[0] + nums[1] == 9`, we return `[0, 1]`.

**Example 2:**
**Input:**
```
nums = [3, 2, 4], target = 6
```
**Output:**
```
[1, 2]
```
**Explanation:** Because `nums[1] + nums[2] == 6`, we return `[1, 2]`.

**Example 3:**
**Input:**
```
nums = [3, 3], target = 6
```
**Output:**
```
[0, 1]
```
**Explanation:** Because `nums[0] + nums[1] == 6`, we return `[0, 1]`.

**Thought Process:**

1.  **Brute-Force Approach:** The simplest idea is to check every possible pair of numbers in the array. This would involve two nested loops: the outer loop picking the first number, and the inner loop picking the second number from the remaining elements. For each pair, check if their sum equals the `target`. If it does, return their indices. This approach has a time complexity of `O(N^2)`.

2.  **Optimizing with a Hash Map:** Can we do better than `O(N^2)`? When we iterate through the array and pick a number `x = nums[i]`, we are essentially looking for another number, let's call it `y`, such that `x + y = target`. This means `y = target - x`.

    If we could quickly find out if `y` exists in the array and what its index is, we could solve this problem much faster. This is where a hash map (dictionary in Python) comes in handy.

3.  **Algorithm Idea with Hash Map:**
    *   We iterate through the array `nums` once.
    *   For each number `nums[i]` (let's call it `current_num`):
        *   Calculate the `complement` needed: `complement = target - current_num`.
        *   Check if this `complement` already exists in our hash map. The hash map will store numbers we've already seen, along with their indices.
        *   If the `complement` *is* in the hash map, it means we've found our two numbers. We return the index of the `complement` (stored in the hash map) and the `current_num`'s index (`i`).
        *   If the `complement` is *not* in the hash map, we add `current_num` and its index `i` to the hash map, so it can be a `complement` for future numbers.

4.  **Why this works:** The problem states there is exactly one solution and we cannot use the same element twice. By storing elements and their indices as we iterate, we ensure that: (a) we find the pair, and (b) if `nums[i]` is used to find its `complement`, that `complement` must have appeared earlier in the array, thus preventing using `nums[i]` itself as its own `complement` (unless the target requires `nums[i]` twice, e.g., `[3,3]` target 6, but then `nums[i]` and `nums[j]` where `i!=j` are distinct elements by index).

**Algorithm:**

1.  Create an empty hash map (dictionary in Python), let's call it `num_map`. This map will store `number -> index` pairs.
2.  Iterate through the `nums` array with index `i` and value `num`:
    a.  Calculate `complement = target - num`.
    b.  Check if `complement` exists as a key in `num_map`:
        i.  If `complement` is in `num_map`, it means we have found the two numbers. Return `[num_map[complement], i]`.
    c.  If `complement` is not in `num_map`, add the current number `num` and its index `i` to the map: `num_map[num] = i`.
3.  (The problem statement guarantees exactly one solution, so the loop will always find a pair and return; we don't need to handle a case where no solution is found.)

````python
def twoSum(nums, target):
    seen_map = {}  # Format: { number: index }
    for i in range(len(nums)):
        num = nums[i]
        complement = target - num
        if complement in seen_map:
            return [seen_map[complement], i]
        seen_map[num] = i
# Test cases
print(f"Two Sum for nums=[2,7,11,15], target=9: {twoSum([2,7,11,15], 9)}") # Expected: [0, 1]
print(f"Two Sum for nums=[3,2,4], target=6: {twoSum([3,2,4], 6)}")       # Expected: [1, 2]
print(f"Two Sum for nums=[3,3], target=6: {twoSum([3,3], 6)}")           # Expected: [0, 1]
print(f"Two Sum for nums=[1,5,9,13], target=14: {twoSum([1,5,9,13], 14)}") # Expected: [1, 2]
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   We iterate through the list `nums` once. `N` is the number of elements in `nums`.
    *   During each iteration, we perform a dictionary lookup (`complement in num_map`) and a dictionary insertion (`num_map[num] = i`). On average, these hash map operations take `O(1)` time.
    *   Therefore, the overall time complexity is `O(N)`.

*   **Space Complexity:** `O(N)`
    *   In the worst case, we might store all `N` numbers and their indices in the `num_map` if no pair is found until the very end.
    *   Therefore, the space complexity is `O(N)`.

### Problem 2: Contains Duplicate

**Problem Statement:**
Given an integer array `nums`, return `true` if any value appears at least twice in the array, and return `false` if every element is distinct.

**Input/Output Example:**

**Example 1:**
**Input:**
```
nums = [1, 2, 3, 1]
```
**Output:**
```
true
```
**Explanation:** The number `1` appears twice.

**Example 2:**
**Input:**
```
nums = [1, 2, 3, 4]
```
**Output:**
```
false
```
**Explanation:** All elements are distinct.

**Example 3:**
**Input:**
```
nums = [1, 1, 1, 3, 3, 4, 3, 2, 4, 2]
```
**Output:**
```
true
```
**Explanation:** Multiple numbers appear more than once (e.g., 1, 3, 4, 2).

**Thought Process:**

1.  **Brute-Force Approach:** We could use nested loops. For each element `nums[i]`, iterate through the rest of the array `nums[j]` (where `j > i`) and check if `nums[i] == nums[j]`. If a match is found, return `true`. If the loops complete without finding any duplicates, return `false`. This approach has a time complexity of `O(N^2)`.

2.  **Optimizing with Sorting:** If we sort the array, any duplicate elements will be adjacent to each other. We can then iterate through the sorted array once, comparing each element with its neighbor. If `nums[i] == nums[i+1]` for any `i`, then a duplicate exists. This approach would have a time complexity dominated by sorting, typically `O(N log N)`.

3.  **Optimizing with a Hash Set:** A hash set (set in Python) stores unique elements and provides very fast (average `O(1)`) lookups for existence. This property makes it ideal for checking for duplicates.

4.  **Algorithm Idea with Hash Set:**
    *   Initialize an empty hash set.
    *   Iterate through each number in the input array `nums`.
    *   For each number, check if it's already present in the hash set.
        *   If it *is* present, it means we've encountered this number before, so it's a duplicate. We can immediately return `true`.
        *   If it's *not* present, add the number to the hash set so we can track its occurrence for subsequent checks.
    *   If we iterate through the entire array without finding any duplicates (i.e., without returning `true`), then all elements are distinct. Return `false`.

**Algorithm:**

1.  Create an empty hash set, let's call it `seen_numbers`.
2.  Iterate through each `num` in the input array `nums`:
    a.  Check if `num` is already in `seen_numbers`:
        i.  If `num` is in `seen_numbers`, return `True`.
    b.  If `num` is not in `seen_numbers`, add `num` to `seen_numbers`.
3.  If the loop completes without returning `True`, it means no duplicates were found. Return `False`.

````python
def containsDuplicate(nums):
    seen = set()  # Create an empty HashSet

    for num in nums:
        if num in seen:  # Instant O(1) lookup
            return True
        seen.add(num)    # Record this number

    return False
# Test cases
print(f"[1,2,3,1] contains duplicate: {containsDuplicate([1,2,3,1])}")             # Expected: True
print(f"[1,2,3,4] contains duplicate: {containsDuplicate([1,2,3,4])}")             # Expected: False
print(f"[1,1,1,3,3,4,3,2,4,2] contains duplicate: {containsDuplicate([1,1,1,3,3,4,3,2,4,2])}") # Expected: True
print(f"[] contains duplicate: {containsDuplicate([])}")                     # Expected: False
print(f"[7] contains duplicate: {containsDuplicate([7])}")                   # Expected: False
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N)`
    *   We iterate through the `nums` array once. `N` is the number of elements in `nums`.
    *   During each iteration, hash set operations (`in` check and `add`) take `O(1)` time on average.
    *   Therefore, the overall time complexity is `O(N)`.

*   **Space Complexity:** `O(N)`
    *   In the worst case, if all elements are distinct, we will store all `N` elements in the `seen_numbers` hash set.
    *   Therefore, the space complexity is `O(N)`.

### Problem 3: Intersection of Two Arrays

**Problem Statement:**
Given two integer arrays `nums1` and `nums2`, return an array of their intersection. Each element in the result must be unique and you may return the result in any order.

**Input/Output Example:**

**Example 1:**
**Input:**
```
nums1 = [1, 2, 2, 1], nums2 = [2, 2]
```
**Output:**
```
[2]
```
**Explanation:** The common elements are 2. Since we need unique elements, the result is `[2]`.

**Example 2:**
**Input:**
```
nums1 = [4, 9, 5], nums2 = [9, 4, 9, 8, 4]
```
**Output:**
```
[9, 4] (or [4, 9], order doesn't matter)
```
**Explanation:** The common elements are 9 and 4. Both appear in both arrays. Note that 9 appears twice in `nums2` but is only included once in the output because elements must be unique.

**Thought Process:**

1.  **Understanding Intersection:** We are looking for elements that are present in *both* `nums1` and `nums2`. The result should contain only *unique* elements.

2.  **Brute-Force Approach:** We could iterate through each element in `nums1`, and for each element, iterate through `nums2` to see if it exists. If it does, add it to a result list, but only if it's not already in the result list (to ensure uniqueness). This would be very inefficient, likely `O(N*M)` where N and M are lengths of arrays, plus the cost of checking for uniqueness in the result list.

3.  **Optimizing with Hash Sets:** Hash sets are perfect for this problem because:
    *   They store only unique elements by nature.
    *   They provide `O(1)` average time complexity for adding elements and checking for existence (`in` operation).

4.  **Algorithm Idea with Hash Sets:**
    *   Convert one of the arrays (e.g., `nums1`) into a hash set. This will automatically handle uniqueness within `nums1` and allow for very fast lookups.
    *   Iterate through the second array (`nums2`). For each element in `nums2`:
        *   Check if this element exists in the hash set created from `nums1`.
        *   If it does, it means it's common to both arrays. Add it to a *new* result hash set (to ensure uniqueness in the final output, as `nums2` might have duplicates).
    *   Finally, convert the result hash set back into a list.

5.  **Alternative using Python's `set` operations:** Python's built-in `set` type has direct methods for set intersection, which is often the most concise and efficient way to solve this specific problem.

**Algorithm (using two hash sets):**

1.  Create an empty hash set, `set1`, and add all unique elements from `nums1` into it.
2.  Create an empty hash set, `intersection_set`, to store the common unique elements.
3.  Iterate through each `num` in `nums2`:
    a.  If `num` is present in `set1`, add `num` to `intersection_set`.
4.  Convert `intersection_set` to a list and return it.

````python
def intersection(nums1, nums2):
    # Step 1: Add all unique elements from nums1 to a hash set
    set1 = set(nums1)
    print(set1)
    # Step 2: Create a set to store the intersection elements
    intersection_set = set()
    # Step 3: Iterate through nums2 and check if elements are in set1
    for num in nums2:
        if num in set1:
            intersection_set.add(num)
    # Step 4: Convert the intersection set to a list and return
    return list(intersection_set)
print(f"Intersection of [1,2,2,1] and [2,2]: {intersection([1,2,2,1], [2,2])}")              # Expected: [2]
print(f"Intersection of [4,9,5] and [9,4,9,8,4]: {intersection([4,9,5], [9,4,9,8,4])}")    # Expected: [9, 4] or [4, 9]
print(f"Intersection of [1,2,3] and [4,5,6]: {intersection([1,2,3], [4,5,6])}")              # Expected: []
print(f"Intersection of [1] and [1]: {intersection([1], [1])}")                            # Expected: [1]
print(f"Intersection of [] and [1,2]: {intersection([], [1,2])}")                          # Expected: []
````

**Complexity Analysis:**

*   **Time Complexity:** `O(N + M)`
    *   Creating `set1` from `nums1` takes `O(N)` time, where `N` is the length of `nums1`.
    *   Iterating through `nums2` takes `O(M)` time, where `M` is the length of `nums2`. Each `in` operation and `add` operation on a hash set takes `O(1)` time on average.
    *   Converting the `intersection_set` to a list takes `O(min(N, M))` time in the worst case (number of unique common elements).
    *   Therefore, the overall time complexity is `O(N + M)`.

*   **Space Complexity:** `O(N + M)`
    *   In the worst case, `set1` will store `N` unique elements and `intersection_set` could store up to `min(N, M)` unique elements.
    *   Therefore, the space complexity is `O(N + M)` (or more precisely, `O(min(N_unique, M_unique))` for the intersection set and `O(N_unique)` for `set1`).

### Problem 1: Merge Sorted Array

**Problem Statement:**
You are given two integer arrays `nums1` and `nums2`, sorted in non-decreasing order, and two integers `m` and `n`, representing the number of elements in `nums1` and `nums2` respectively.

Merge `nums1` and `nums2` into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array `nums1`. To accommodate this, `nums1` has a length of `m + n`, where the first `m` elements denote the elements that should be merged, and the last `n` elements are `0`s and should be ignored. `nums2` has a length of `n`.

**Input/Output Example:**

**Example 1:**
**Input:**
```
nums1 = [1, 2, 3, 0, 0, 0], m = 3, nums2 = [2, 5, 6], n = 3
```
**Output:**
```
[1, 2, 2, 3, 5, 6]
```
**Explanation:** The arrays we are merging are `[1, 2, 3]` and `[2, 5, 6]`. The result of the merge is `[1, 2, 2, 3, 5, 6]`.

**Example 2:**
**Input:**
```
nums1 = [1], m = 1, nums2 = [], n = 0
```
**Output:**
```
[1]
```
**Explanation:** The arrays we are merging are `[1]` and `[]`. The result of the merge is `[1]`.

**Example 3:**
**Input:**
```
nums1 = [0], m = 0, nums2 = [1], n = 1
```
**Output:**
```
[1]
```
**Explanation:** The arrays we are merging are `[]` and `[1]`. The result of the merge is `[1]`. Note that because `m = 0`, `nums1` initially contains only `0` at index 0. This `0` is just a placeholder and should be overwritten.

**Thought Process:**

1.  **Direct Merging and Sorting (Naive):** The simplest but often inefficient approach would be to copy all elements from `nums2` into the empty slots of `nums1`, and then sort `nums1`. This would be `O((m+n) log (m+n))` due to sorting, which might not be optimal given that both arrays are already sorted.

2.  **Two Pointers (From Beginning):** Since both arrays are sorted, we can use two pointers, one for `nums1` (say `p1`) and one for `nums2` (say `p2`). We can compare `nums1[p1]` and `nums2[p2]` and place the smaller one into a new auxiliary array. The issue here is that the problem requires modifying `nums1` *in-place*. If we try to merge into `nums1` from the beginning, we would overwrite elements of `nums1` that haven't been processed yet.

3.  **Two Pointers (From End - Optimal):** The key insight for an in-place merge of two sorted arrays when one has extra space at the end is to work *backwards*. This way, we fill the empty slots in `nums1` without overwriting elements we still need to compare.
    *   We need three pointers:
        *   `p1`: Points to the last actual element of `nums1` (initially `m-1`).
        *   `p2`: Points to the last element of `nums2` (initially `n-1`).
        *   `write_idx`: Points to the last position in `nums1` where an element can be placed (initially `m + n - 1`).
    *   We compare `nums1[p1]` and `nums2[p2]` and place the larger one at `nums1[write_idx]`. Then we decrement the pointer of the element that was moved and `write_idx`.
    *   We continue this process until one of the pointers (`p1` or `p2`) goes below 0.
    *   **What if `nums2` still has elements left?** If `p2` is still `>= 0` but `p1` has gone below 0, it means all remaining elements in `nums2` are smaller than all elements already placed in `nums1`. We just copy these remaining `nums2` elements directly into the front of `nums1`.

**Algorithm (Two Pointers from End):**

1.  Initialize three pointers:
    *   `p1 = m - 1` (last element of the meaningful part of `nums1`)
    *   `p2 = n - 1` (last element of `nums2`)
    *   `write_idx = m + n - 1` (last position in `nums1`)
2.  While `p1 >= 0` and `p2 >= 0`:
    a.  If `nums1[p1] > nums2[p2]`:
        i.  Place `nums1[p1]` at `nums1[write_idx]`.
        ii. Decrement `p1`.
    b.  Else (`nums2[p2] >= nums1[p1]`):
        i.  Place `nums2[p2]` at `nums1[write_idx]`.
        ii. Decrement `p2`.
    c.  Decrement `write_idx`.
3.  After the loop, if there are still elements left in `nums2` (i.e., `p2 >= 0`):
    a.  Copy the remaining elements from `nums2` (from `p2` down to 0) into `nums1` starting from `write_idx` down to 0.
    b.  This can be done with another `while p2 >= 0` loop: `nums1[write_idx] = nums2[p2]`, then decrement `p2` and `write_idx`.

````python
def merge(nums1, m, nums2, n):
#Merges nums2 into nums1 in-place. Both arrays are sorted non-decreasingly.nums1 has a length of m + n.

    last =m+n-1
    print(last)
    while m>0 and n>0:
      if nums1[m-1]>nums2[n-1]:
        nums1[last]=nums1[m-1]
        m-=1
        print(m)
      else:
        nums1[last]=nums2[n-1]
        n-=1
        print(n)
      last-=1
    while n>0:
      nums1[last]=nums2[n-1]
      n-=1
      last-=1

nums1_4 = [1, 2, 3, 0, 0, 0]
m_4 = 3
nums2_4 = [2,5,6]
n_4 = 3
merge(nums1_4, m_4, nums2_4, n_4)
print(f"Merged array for nums1=[4,5,6,0,0,0], m=3, nums2=[1,2,3], n=3: {nums1_4}") # Expected: [1, 2, 3, 4, 5, 6]
````

**Complexity Analysis:**

*   **Time Complexity:** `O(m + n)`
    *   We iterate through the arrays from the end, effectively visiting each element of `nums1` (up to `m` elements) and `nums2` (up to `n` elements) at most once.
    *   Each comparison and assignment operation takes `O(1)` time.
    *   Therefore, the total time complexity is linear, proportional to the total number of elements in both arrays.

*   **Space Complexity:** `O(1)`
    *   We are performing the merge in-place in `nums1` and only using a few extra pointers (variables).
    *   No auxiliary data structures that grow with input size are used.

````python
def arrayPairSum(nums):
    # Sort the array inline to pair up closest values
    nums.sort()

    max_sum = 0

    # Iterate through the array taking steps of 2
    for i in range(0, len(nums), 2):
        max_sum += nums[i]

    return max_sum
print(arrayPairSum([1, 4, 3, 2,1,3,5]))
````

**Problem 3: Height Checker**
Understanding the Goal
A school wants to take a group photo, and students are asked to stand in non-decreasing order by their heights. You are given an array heights representing the current order students are standing in. Return the total number of indices where heights[i] does not match the correct expected height order.

Example: heights = [1, 1, 4, 2, 1, 3]

Output: 3

Pattern Recognition
To determine how many elements are out of place, you need an absolute reference point showing what the array should look like when perfectly organized. This matches the Reference Array Comparison Pattern.

The Logic & Solution Approach
Create a separate copy of the array and sort it. This copy represents the perfect expected order.

Initialize a tracking counter count = 0.

Loop through the arrays from index 0 to the end, comparing the element in the original heights array against the element at the same index in the expected array.

If heights[i] != expected[i], it means a student is out of position. Increment your counter by 1.

Return the final counter value.

````python
def heightChecker(heights):
    # Create a brand new sorted version as our baseline reference
    expected = sorted(heights)

    mismatch_count = 0

    # Compare original alignment vs expected alignment side-by-side
    for i in range(len(heights)):
        if heights[i] != expected[i]:
            mismatch_count += 1

    return mismatch_count
print(heightChecker([1, 1, 4, 2, 1, 3]))
````
