Tree is a non-linear, hierarchical data structure. Unlike arrays, linked lists, stacks, and queues (which are linear), trees store data in a layered, parent-child relationship.

Think of it like an organizational chart or a family tree, just inverted—with the "root" at the very top.

**Core Tree Terminology**


Node: The fundamental building block of a tree. It contains a data value and

links (pointers) to other nodes.

Root: The topmost node in a tree. It has no parent. A tree can have only one root.

Edge: The connecting link or pointer between two nodes.

Parent: A node that has an edge linking to another node below it.

Child: A node connected to a parent node above it.

Leaf: A node that has no children (the endpoints of the tree).

Subtree: Any node and all of its descendants can be viewed as a smaller tree within the larger tree.

Height of a Node: The number of edges on the longest path from that node to a leaf.

Depth of a Node: The number of edges from the root to that node.

````python
#1. Trees - Maximum Depth of Trees

# Let's start by defining what a 'Node' in a tree looks like.
# Think of a tree as an upside-down plant: it has a 'root' (the top) and 'branches' that lead to 'leaves' (the bottom).
# Each point where a branch splits or ends is called a 'Node'.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        # 'val' is the value stored in this node. It could be any data, like a number or a name.
        self.val = val
        # 'left' is a pointer to the 'child' node on its left side.
        # If there's no left child, it will be 'None'.
        self.left = left
        # 'right' is a pointer to the 'child' node on its right side.
        # If there's no right child, it will be 'None'.
        self.right = right

# Now, let's define a function to calculate the 'Maximum Depth' of a tree.
# The depth of a tree is the number of nodes along the longest path from the root node down to the farthest leaf node.
# A 'leaf node' is a node that has no children.

def maxDepth(root: TreeNode) -> int:
    # Step 1: Handle the 'base case'.
    # What if the tree is empty? If 'root' is None, it means there are no nodes.
    # An empty tree has a depth of 0.
    if root is None:
        return 0

    # Step 2: Recursively find the depth of the left subtree.
    # We call 'maxDepth' on the 'left' child. This is like asking: "What's the deepest branch on my left?"
    left_depth = maxDepth(root.left)

    # Step 3: Recursively find the depth of the right subtree.
    # Similarly, we ask: "What's the deepest branch on my right?"
    right_depth = maxDepth(root.right)

    # Step 4: Combine the results.
    # The maximum depth of the entire tree (starting from the current 'root' node)
    # will be 1 (for the current 'root' node itself) plus the maximum of the depths
    # we found in its left and right subtrees.
    return 1 + max(left_depth, right_depth)

# --- Let's test our function with an example! ---

# Example 1: A simple tree with a root and two children
#     3
#    / \
#   9  20
#     /  \
#    15   7

# First, create the leaf nodes (nodes with no children)
node15 = TreeNode(15)
node7 = TreeNode(7)

# Then, create the next level nodes, linking them to their children
node9 = TreeNode(9)
node20 = TreeNode(20, node15, node7) # node20 has node15 as left child, node7 as right child

# Finally, create the root node
root = TreeNode(3, node9, node20) # root has node9 as left child, node20 as right child

print("Example 1: Max depth of the tree:", maxDepth(root))
# Expected output: 3 (Path: 3 -> 20 -> 15 or 3 -> 20 -> 7)

# Example 2: A very simple tree with just a root
#   1
root2 = TreeNode(1)
print("Example 2: Max depth of a single-node tree:", maxDepth(root2))
# Expected output: 1

# Example 3: An empty tree
root3 = None
print("Example 3: Max depth of an empty tree:", maxDepth(root3))
# Expected output: 0
````

## Depth-First Search (DFS) and Breadth-First Search (BFS)

DFS and BFS are two fundamental algorithms used for traversing or searching tree or graph data structures. They differ in the order in which they visit the nodes.

---

### Depth-First Search (DFS)

**Concept:**
Depth-First Search explores as far as possible along each branch before backtracking. Think of it like exploring a maze: you go down one path as far as you can, and if it's a dead end, you backtrack and try another path.

**How it works (for a tree):**
1.  Start at the root (or any arbitrary node).
2.  Explore one of its children (say, the left child) completely before moving to the next sibling (right child).
3.  It goes as deep as possible into each branch before moving to the next branch.

**Analogy:** Imagine you're searching for a book in a library. With DFS, you pick a bookshelf, search it from top to bottom, then move to the next bookshelf only after you've thoroughly checked the current one.

**Data Structure Used:** Typically implemented using a **Stack** (explicitly or implicitly via recursion).

**Common Applications:**
*   Finding a path between two nodes.
*   Topological sorting.
*   Detecting cycles in a graph.
*   Solving puzzles with a single solution (e.g., mazes).

**Traversal Orders (for trees):**
DFS can yield different traversal orders for trees:
*   **Pre-order traversal:** Visit root, then left subtree, then right subtree.
*   **In-order traversal:** Visit left subtree, then root, then right subtree (common for binary search trees).
*   **Post-order traversal:** Visit left subtree, then right subtree, then root.

---

### Breadth-First Search (BFS)

**Concept:**
Breadth-First Search explores all the nodes at the current depth level before moving on to the nodes at the next depth level. It explores 'layer by layer'.

**How it works (for a tree):**
1.  Start at the root.
2.  Visit all its immediate children.
3.  Then, for each of those children, visit all *their* immediate children, and so on.

**Analogy:** Using the library example, with BFS, you'd scan all the books on the first shelf, then all the books on the second shelf, then all on the third, and so on, until you find your book.

**Data Structure Used:** Typically implemented using a **Queue**.

**Common Applications:**
*   Finding the shortest path between two nodes in an unweighted graph (since it explores level by level, the first time it finds the target, it's the shortest path).
*   Finding all connected components in a graph.
*   Web crawlers (exploring pages linked from a given page).
*   Level order traversal of a tree.

---

### Key Differences Summarized:

| Feature            | Depth-First Search (DFS)             | Breadth-First Search (BFS)             |
| :----------------- | :----------------------------------- | :------------------------------------- |
| **Order**          | Explores deeply first                | Explores broadly (level by level) first |
| **Data Structure** | Stack (or recursion call stack)      | Queue                                  |
| **Goal**           | Good for path finding, cycle detection, topological sort | Good for shortest path on unweighted graphs, finding all nodes at a given depth |
| **Memory**         | Can be less memory-intensive for deep, narrow graphs | Can be more memory-intensive for wide graphs (stores all nodes at a level) |
| **Path Finding**   | Finds *a* path                       | Finds the *shortest* path (in terms of number of edges) |


## Binary Tree Traversals: Inorder, Preorder, and Postorder

Binary tree traversal is the process of visiting each node in the tree exactly once in a specific order. There are three common ways to traverse a binary tree using Depth-First Search (DFS) principles:

1.  **Inorder Traversal**
2.  **Preorder Traversal**
3.  **Postorder Traversal**

To understand these, let's consider a generic Binary Tree node definition (similar to the `TreeNode` class we defined earlier):

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

Let's use the following example tree for demonstration:

```
        1
       / \
      2   3
     / \ / \
    4  5 6  7
```

---

### 1. Inorder Traversal (Left -> Root -> Right)

**Concept:** Inorder traversal visits the left subtree first, then the root node, and finally the right subtree.

**Process:**
1.  Recursively traverse the left subtree.
2.  Visit the root node.
3.  Recursively traverse the right subtree.

**Analogy:** If the tree represents a Binary Search Tree (where left children are smaller than the root, and right children are larger), an inorder traversal will visit the nodes in **ascending order** of their values.

**Example Traversal for our tree:**

-   From Root (1):
    -   Go Left to (2)
        -   Go Left to (4) -> **Visit 4**
        -   Return to (2) -> **Visit 2**
        -   Go Right to (5) -> **Visit 5**
    -   Return to (1) -> **Visit 1**
    -   Go Right to (3)
        -   Go Left to (6) -> **Visit 6**
        -   Return to (3) -> **Visit 3**
        -   Go Right to (7) -> **Visit 7**

**Output:** `4, 2, 5, 1, 6, 3, 7`

**Python Implementation:**
```python
def inorder_traversal(root: TreeNode):
    result = []
    if root:
        result.extend(inorder_traversal(root.left)) # Traverse left
        result.append(root.val)                     # Visit root
        result.extend(inorder_traversal(root.right)) # Traverse right
    return result
```

---

### 2. Preorder Traversal (Root -> Left -> Right)

**Concept:** Preorder traversal visits the root node first, then recursively traverses the left subtree, and finally recursively traverses the right subtree.

**Process:**
1.  Visit the root node.
2.  Recursively traverse the left subtree.
3.  Recursively traverse the right subtree.

**Analogy:** Think of this as creating a **copy** of the tree structure. When you visit the root first, you're essentially recording the 'blueprint' of the tree's hierarchy.

**Example Traversal for our tree:**

-   From Root (1) -> **Visit 1**
    -   Go Left to (2) -> **Visit 2**
        -   Go Left to (4) -> **Visit 4**
        -   Return to (2)
        -   Go Right to (5) -> **Visit 5**
    -   Return to (1)
    -   Go Right to (3) -> **Visit 3**
        -   Go Left to (6) -> **Visit 6**
        -   Return to (3)
        -   Go Right to (7) -> **Visit 7**

**Output:** `1, 2, 4, 5, 3, 6, 7`

**Python Implementation:**
```python
def preorder_traversal(root: TreeNode):
    result = []
    if root:
        result.append(root.val)                     # Visit root
        result.extend(preorder_traversal(root.left))  # Traverse left
        result.extend(preorder_traversal(root.right)) # Traverse right
    return result
```

---

### 3. Postorder Traversal (Left -> Right -> Root)

**Concept:** Postorder traversal recursively traverses the left subtree, then recursively traverses the right subtree, and finally visits the root node.

**Process:**
1.  Recursively traverse the left subtree.
2.  Recursively traverse the right subtree.
3.  Visit the root node.

**Analogy:** This traversal is often used to **delete** a tree. By deleting the children first and then the parent, you ensure that no references to deleted nodes remain.

**Example Traversal for our tree:**

-   From Root (1):
    -   Go Left to (2)
        -   Go Left to (4) -> **Visit 4**
        -   Return to (2)
        -   Go Right to (5) -> **Visit 5**
        -   Return to (2) -> **Visit 2**
    -   Return to (1)
    -   Go Right to (3)
        -   Go Left to (6) -> **Visit 6**
        -   Return to (3)
        -   Go Right to (7) -> **Visit 7**
        -   Return to (3) -> **Visit 3**
    -   Return to (1) -> **Visit 1**

**Output:** `4, 5, 2, 6, 7, 3, 1`

**Python Implementation:**
```python
def postorder_traversal(root: TreeNode):
    result = []
    if root:
        result.extend(postorder_traversal(root.left))  # Traverse left
        result.extend(postorder_traversal(root.right)) # Traverse right
        result.append(root.val)                     # Visit root
    return result
```

````python
#1.Binary tree inorder traversal
class TreeNode():
  def __init__(self,val=0,left=None,right=None):
    self.val=val
    self.left=left
    self.right=right
class Solution:
  def inordertraversal(self,root:TreeNode)->list[int]:
    result=[]
    def dfs(node):
      # The base case should check if the current 'node' is None, not the global 'root'.
      if node is None:
        return
      dfs(node.left)
      result.append(node.val)
      dfs(node.right)
    dfs(root)
    return result

# Example Usage:
node15=TreeNode(15)
node7=TreeNode(7)
node9=TreeNode(9)
node20=TreeNode(20,node15,node7)
root=TreeNode(3,node9,node20)
print(Solution().inordertraversal(root))

# Expected output for the given tree structure:
#       3
#      / \
#     9  20
#       /  \
#      15   7
# Inorder traversal: Left -> Root -> Right
````
